## Розділ 4. Групування, агрегація, статистика
**Мета:** навчитися використовувати `groupby`, агрегації, зведені таблиці.

**Завдання 27–35:**
27. Створи DataFrame про продажі з колонками: `"Product"`, `"Category"`, `"Price"`, `"Quantity"` (10–15 рядків).  
28. Знайди загальний дохід (`Price * Quantity`) і додай його як новий стовпець `"Revenue"`.  
29. Порахуйте загальний дохід для кожної категорії (`groupby("Category")["Revenue"].sum()`).  
30. Знайди середню ціну по кожному `"Product"`.  
31. Визнач найпопулярніший товар (за сумарною кількістю продажів).  
32. Знайди мінімальний і максимальний дохід по категорії.  
33. Зроби зведену таблицю (`pivot_table`) з категоріями по рядках і середнім доходом по стовпцях.  
34. Відсортуй категорії за середнім доходом.  
35. Додай стовпець `"Profit" = "Revenue" * 0.25` і знайди середній прибуток по категоріях.  

````python
# Розв'язання для завдань 27–35
# Підказка: можеш згенерувати випадкові дані за допомогою np.random.

# task 27. Створи DataFrame про продажі з колонками: `"Product"`, `"Category"`, `"Price"`, `"Quantity"` (10–15 рядків). 
products = ["Laptop", "Phone", "Tablet", "Monitor", "Keyboard", "Mouse", "Headphones", "Printer", "Camera", "Speaker"]
categories = ["Electronics", "Accessories", "Peripherals"]
n = 15
df_sales = pd.DataFrame({
    "Product": np.random.choice(products, size=n),
    "Category": np.random.choice(categories, n),
    "Price": np.round(np.random.uniform(50, 1500, n), 2),
    "Quantity": np.random.randint(1, 20, n)
})
df_sales
````

````python
# task 28. Знайди загальний дохід (`Price * Quantity`) і додай його як новий стовпець `"Revenue"`. 
df_sales["Revenue"] = df_sales["Price"] * df_sales["Quantity"]
df_sales
````

````python
# task 29. Порахуйте загальний дохід для кожної категорії (`groupby("Category")["Revenue"].sum()`).  
df_sales.groupby("Category")["Revenue"].sum()
````

````python
# task 30. Знайди середню ціну по кожному `"Product"`. 
df_sales.groupby("Product")["Price"].mean()
````

````python
# task 31. Визнач найпопулярніший товар (за сумарною кількістю продажів).
df_sales.groupby("Product")["Quantity"].sum().idxmax()
````

````python
# task 32. Знайди мінімальний і максимальний дохід по категорії. 
df_sales.groupby("Category")["Revenue"].agg(["min", "max"])
````

````python
# task 33. Зроби зведену таблицю (`pivot_table`) з категоріями по рядках і середнім доходом по стовпцях.  
pivot_tab = df_sales.pivot_table(
    index="Category",
    values="Revenue",
    aggfunc="mean"
    )
pivot_tab
````

````python
# task 34. Відсортуй категорії за середнім доходом.
pivot_tab.sort_values(by="Revenue", ascending=False)
````

````python
# task 35. Додай стовпець `"Profit" = "Revenue" * 0.25` і знайди середній прибуток по категоріях.
df_sales["Profit"] = df_sales["Revenue"] * 0.25
df_sales.groupby("Category")["Profit"].mean()
````

````python
# Розв'язання для завдань 36–40
# task 36. Створи два DataFrame: `customers` і `orders`:
#    - `customers`: `CustomerID`, `Name`, `City`  
#    - `orders`: `OrderID`, `CustomerID`, `Amount`  
#    З’єднай їх за `CustomerID` (`pd.merge`). 
customers = pd.DataFrame({
    "CustomerID": [1, 2, 3, 4, 5],
    "Name": ["Anna", "Victor", "Elsa", "Ted", "Olga"],
    "City": ["Lviv", "Kyiv", "Paris", "London", "Kil"]
})
orders = pd.DataFrame({
    "OrderID": [101, 102, 103, 104, 105],
    "CustomerID": [1, 2, 1, 3, 4],
    "Amount": [250, 150, 300, 200, 400]
})
pd.merge(orders, customers, on="CustomerID", how="inner")
````

````python
# task 37. Об’єднай дві таблиці вертикально (`pd.concat`).
pd.concat([orders, customers])
````

````python
# task 38. Знайди клієнтів, які **не мають** замовлень (left join + `NaN` у `"Amount"`).
merged = pd.merge(customers, orders, on="CustomerID", how="left")
merged[merged["Amount"].isna()]
````

````python
# task 39. Збережи результат у CSV-файл (`.to_csv("customers_orders.csv", index=False)`) і потім зчитай його назад (`pd.read_csv`). 
merged.to_csv("customers_orders.csv", index=False)
read_df = pd.read_csv("customers_orders.csv")
read_df
````

````python
# Розв'язання для завдань 61–70
# task 61. Створи або зчитай з CSV DataFrame з даними про продажі: дата, товар, категорія, ціна, кількість, місто. 
df_sales2 = pd.DataFrame({
    "Date": pd.to_datetime(["2025-12-15", "2025-01-10", "2025-02-05", "2025-03-20", "2025-04-12", "2025-11-25", "2025-06-30", "2025-07-18"]),
    "Product": ["Laptop", "Phone", "Tablet", "Monitor", "Keyboard", "Mouse", "Camera", "Headphones"],
    "Category": ["Electronics", "Electronics", "Electronics", "Electronics", "Peripherals", "Peripherals", "Electronics", "Accessories"],
    "Price": [1200, 800, 500, 300, 100, 50, 600, 150],
    "Quantity": [5, 10, 8, 12, 20, 25, 7, 15],
    "City": ["Kyiv", "Lviv", "Kyiv", "Odessa", "Lviv", "Odessa", "Kyiv", "Lviv"]
})
df_sales2
````

````python
# task 62. Порахуйте загальний дохід по місяцях і побудуйте лінійний графік.
df_sales2["Revenue"] = df_sales2["Price"] * df_sales2["Quantity"]
df_sales2["Month"] = df_sales2["Date"].dt.month
monthly_revenue = df_sales2.groupby("Month")["Revenue"].sum()
monthly_revenue.plot(kind="line", marker="o")
plt.title("Monthly Revenue in 2025")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(range(1,13))  # Показує всі місяці
plt.grid(True)
plt.show()
````

````python
# task 63. Визнач топ-3 товари за продажами (за `Revenue`) і побудуй для них стовпчиковий графік.
top3_products = df_sales2.nlargest(3, "Revenue")
plt.bar(top3_products["Product"], top3_products["Revenue"])
plt.title("Top 3 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
````

````python
# task 64. Побудуй порівняння доходів по містах (стовпчиковий графік).  
city_revenue = df_sales2.groupby("City")["Revenue"].sum().sort_values(ascending=False)
plt.bar(city_revenue.index, city_revenue.values, color="lightgreen")
plt.title("Revenue by City")
plt.xlabel("City")
plt.ylabel("Revenue")
plt.grid(axis="y")
plt.show()
````

````python
# task 65. Обчисли кореляцію між числовими стовпцями (`df.corr()`) і побудуй `heatmap` (через `plt.imshow` або іншу бібліотеку). 
plt.imshow(df_sales2[["Price", "Quantity"]].corr())
plt.title("Correlation Heatmap")
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.show()
plt.figure(figsize=(6,5))
sns.heatmap(df_sales2[["Price", "Quantity", "Revenue"]].corr())
plt.title("Correlation Heatmap")
plt.show()
````

````python
# task 66. Побудуй `subplot` із двома графіками: лінійний (Revenue по місяцях) і стовпчиковий (Revenue по категоріях). 
monthly_revenue = df_sales2.groupby("Month")["Revenue"].sum()
category_revenue = df_sales2.groupby("Category")["Revenue"].sum()
plt.figure(figsize=(14,5))
plt.subplot(1, 2, 1)
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o')
plt.title("Revenue by Month")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.subplot(1, 2, 2)
plt.bar(category_revenue.index, category_revenue.values)
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.grid(axis="y")

plt.tight_layout()
plt.show()
````

````python
# task 67. Створи кілька серій даних (наприклад, Revenue за 2 роки) і побудуй їх разом на одному графіку з легендою.
monthly_category_revenue = df_sales2.pivot_table(index="Month", columns="Category", values="Revenue", aggfunc="sum").fillna(0)
monthly_category_revenue.plot(marker="o", figsize=(10,5))
plt.title("Monthly Revenue by Category")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.legend(title="Category")
plt.show()
````

````python
# task 68. Побудуй графік із вторинною віссю (`twinx`): стовпчики для Revenue і лінія для середньої ціни.  
monthly_revenue = df_sales2.groupby("Month")["Revenue"].sum()
monthly_avg_price = df_sales2.groupby("Month")["Price"].mean()
plt.figure(figsize=(10,5))
plt.bar(monthly_revenue.index, monthly_revenue.values, color="blue")
plt.xlabel("Month")
plt.ylabel("Revenue", color="blue")
plt.twinx()
plt.plot(monthly_avg_price.index, monthly_avg_price.values, color="red", marker="o", linewidth=2)
plt.ylabel("Average Price", color="red")
plt.title("Monthly Revenue and Average Price")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
````
